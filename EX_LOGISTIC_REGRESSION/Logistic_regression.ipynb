```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score, mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

**LOGISTIC REGRESSION - GRADIENT DESCENDENT**

```python
from sklearn.datasets import load_breast_cancer
```

```python
dataset = load_breast_cancer()
```

```python
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
```

```python
df['target'] = dataset.target
```

```python
X = df.iloc[:,:-1]
Y = df.iloc[:,-1]
```

```python
class Logistic_regression:
    def __init__(self,learning_rate, number_iteractions):
        self.learning_rate = learning_rate
        self.number_iteractions = number_iteractions
        self.w = None
        self.b = None


    def fit(self, X, Y):
        self.m , self.n = X.shape
        self.w = np.zeros(self.n)
        self.b = 0
        self.X = X
        self.Y = Y

        for i in range(self.number_iteractions):
            self.update_weights()

    def update_weights(self):    
        Z = np.dot(self.X,self.w) + self.b
        Y_hat = 1 /(1+np.exp(-Z))


        dw = (1/self.m) * np.dot(self.X.T,(Y_hat-self.Y))
        db = (1/self.m) * np.sum(Y_hat-self.Y)
        
        self.w = self.w - self.learning_rate * dw
        self.b = self.b - self.learning_rate * db

    def predict(self,Xtest):

        Z = np.dot(Xtest,self.w) + self.b
        y_pred = 1 /(1+np.exp(-Z))
        y_pred = np.where(y_pred > 0.5,1,0)
        return y_pred
```

```python
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X,Y, test_size=0.3, random_state=42)
```

```python
model = Logistic_regression(learning_rate=0.00001,number_iteractions=200)
```

```python
model.fit(Xtrain,Ytrain)
```

```python
y_pred = model.predict(Xtest)
```

```python
accuracy = accuracy_score(y_pred,Ytest)
mse = mean_squared_error(y_pred,Ytest)
```
